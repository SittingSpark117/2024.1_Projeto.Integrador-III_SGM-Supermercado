```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()
```

```python
df = pd.read_csv("supermarket_sales.csv")
df.head(10)
```

```python
df = df.rename(columns={
    "Invoice ID": "ID da Fatura",
    "Branch": "Filial",
    "City": "Cidade",
    "Customer type": "Tipo de Cliente",
    "Gender": "Gênero",
    "Product line": "Linha de Produto",
    "Unit price": "Preço Unitário",
    "Quantity": "Quantidade",
    "Tax 5%": "Imposto 5%",
    "Total": "Total",
    "Date": "Data",
    "Time": "Hora",
    "Payment": "Pagamento",
    "cogs": "COGS",
    "gross margin percentage": "Margem Bruta (%)",
    "gross income": "Renda Bruta",
    "Rating": "Avaliação"
})

# Salvar o novo arquivo CSV
# df.to_csv('dados_t.csv', index=False)
```

```python
df.head(10)
```

```python
df["Data"] = pd.to_datetime(df["Data"])
data_minima = df["Data"].min()
data_maxima = df["Data"].max()
print("Data mínima:", data_minima)
print("Data máxima:", data_maxima)
```

```python
# tabela lojas
valores_unicos = df.drop_duplicates(subset=["Cidade", "Filial"])
lojas = valores_unicos[["Cidade", "Filial"]]
nomes_ficticios = ["Loja A", "Loja B", "Loja C"]
lojas["Nome da Loja"] = nomes_ficticios
#lojas
lojas.to_csv('lojas.csv', index=False)
```

```python
import random
#df = df.drop("Nome", axis=1)
nomes_ficticios = {
    ("Member", "Female"): ["Maria", "Ana", "Clara", "Patrícia", "Camila", "Aline", "Vanessa"],
    ("Normal", "Female"): ["Joana", "Carla", "Laura", "Mariana", "Renata", "Isabela", "Larissa"],
    ("Normal", "Male"): ["Pedro", "Lucas", "Mateus", "Gabriel", "Bruno", "André", "Ricardo"],
    ("Member", "Male"): ["João", "Rafael", "Gustavo", "Fernando", "Diego", "Marcelo", "Leonardo"]
}

def nome_aleatorio(tipo_cliente, genero):
    nomes = nomes_ficticios.get((tipo_cliente, genero))
    if nomes:
        return random.choice(nomes)
    else:
        return "Desconhecido"


df["Nome"] = df.apply(lambda row: nome_aleatorio(
    row["Tipo de Cliente"], row["Gênero"]), axis=1)


df.to_csv('dados_t.csv', index=False)
```

```python
# tabela clientes
valores_unicos = df.drop_duplicates(
    subset=["Tipo de Cliente", "Gênero", "Nome"])
clientes = valores_unicos[["Tipo de Cliente", "Gênero", "Nome"]]
# clientes
clientes.to_csv('clientes.csv', index=False)
```

```python
# tabela produtos
valores_unicos = df.drop_duplicates(
    subset=["Linha de Produto", "Preço Unitário"])
produtos = valores_unicos[["Linha de Produto", "Preço Unitário"]]
# produtos
produtos.to_csv('produtos.csv', index=False)
```

```python
# tabela pagamento
valores_unicos = df.drop_duplicates(
    subset=["Pagamento"])
pagamento = valores_unicos[["Pagamento"]]
# pagamento
pagamento.to_csv('pagamento.csv', index=False)
```
